```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
v=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
```

```python
v=v.drop(columns=["id","Unnamed: 32"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(v.iloc[:,1:],v.iloc[:,0],test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
std=StandardScaler()
```

```python
x_train,x_test=std.fit_transform(x_train),std.fit_transform(x_test)
```

```python
label=LabelEncoder()
```

```python
y_train,y_test=label.fit_transform(y_train),label.fit_transform(y_test)
```

```python
y_train
```

```python
import torch as t
```

```python
import numpy as np
```

```python
x_train_t,x_test_t,y_train_t,y_test_t=t.from_numpy(x_train.astype(np.float32)),t.from_numpy(x_test.astype(np.float32)),t.from_numpy(y_train.astype(np.float32)),t.from_numpy(y_test.astype(np.float32))
```

```python
x_train_t
```

```python
import torch as t
```

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomDataset(Dataset):
  def __init__(self,features,labels):
    self.features=features
    self.labels=labels
  def __len__(self):
    return (self.features.shape[0])
  def __getitem__(self,idx):
      return self.features[idx],self.labels[idx]
```

```python

dataset_train_t=CustomDataset(x_train_t,y_train_t)
dataset_train_t[0]
```

```python
dataset_test_t=CustomDataset(x_test_t,y_test_t)
```

```python
dataset_train_t_loader=DataLoader(dataset_train_t,batch_size=32,shuffle=True)
```

```python
dataset_test_t_loader=DataLoader(dataset_test_t,batch_size=32,shuffle=True)
```

```python
import torch.nn as nn
```

```python
class Simple_nn(nn.Module):
  def __init__(self,features):
    super().__init__()
    self.container=nn.Sequential(nn.Linear(features,1),nn.Sigmoid())
  def forward(self,x):
    z=self.container(x)
    return z
```

```python
#x_train_t,x_test_t,y_train_t,y_test_t=x_train_t.float(),x_test_t.float(),y_train_t.float(),y_test_t.float()
```

```python
model=Simple_nn(x_train_t.shape[1])
```

```python
model(x_train_t)
```

```python
loss1=nn.BCELoss()
```

```python
#ptimmizer
optimizer=t.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs=25
```

```python
batch_size=32
```

```python
for i in range(epochs):
  for batch_features,batch_label in dataset_train_t_loader:
    #forward_pass
    y_pred=model(batch_features)
    #loss
    loss_=loss1(y_pred,batch_label.reshape(-1,1))
    #clear gradient
    optimizer.zero_grad()
    #backward()
    loss_.backward()
    #update weights
    optimizer.step()
    print(loss_)
```

```python
model(x_train_t)
```

```python
model.eval()
accuracy1=[]
#evaluation
with t.no_grad():
  for batch_features,batch_label in dataset_test_t_loader:
    #forward
    y_pred=model(batch_features)
    y_pred=(y_pred>0.8).float()
    #accuracy
    batch_accuracy=((y_pred==batch_label.reshape(-1,1)).float()).mean()
    accuracy1.append(batch_accuracy)
```

```python
accuracy1
```

```python
#overall_accuracy
overall_accuracy=sum(accuracy1)/len(accuracy1)
print(f'overall_accuracy:{overall_accuracy}')
```
